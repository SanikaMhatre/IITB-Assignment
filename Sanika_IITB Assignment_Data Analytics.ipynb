```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Pie chart of responses from college students 

```python
#importing the dataset set 
#add the path where your csv file is located (College Students 2.csv)
college_students = pd.read_csv(r"C:\Users\sanik\OneDrive\Desktop\Sanika\excel\IITB\College Students 2.csv")
college_students                           
```

```python
#strip any extra whitespaces in the columns
college_students.columns = college_students.columns.str.strip()
```

```python
#plot the pie chart
college_students.plot.pie(y = 'Frequency', labels = college_students['Survey Response'], autopct = '%1.0f', textprops={'fontsize': 10},  startangle = -90,  figsize = (8, 6), labeldistance=1.15)

#add title to the chart and adjust the position of the legend
plt.title('Pie Chart for responses from College Students')
plt.legend(bbox_to_anchor=(0.5, -0.2), loc="center left")
plt.axis('off')
plt.show()
```

# Bar chart for deaths in UK during Pandemic

```python
#importing the dataset set 
#add the path where your csv file is located (UK Data_injection is the Pandemic.csv)
uk_deaths = pd.read_csv(r"C:\Users\sanik\OneDrive\Desktop\Sanika\excel\IITB\UK Data_injection is the Pandemic.csv")
uk_deaths
```

```python
#set Years as index
uk_deaths.set_index('Years', inplace=True)
```

```python
#plot the bar chart
plt_bar_chart = uk_deaths.plot.barh(figsize=(10, 6), legend=True)
plt.title('Bar Chart for deaths in UK during Pandemic')
plt.xlabel('Deaths')
plt.ylabel('Year')

plt.show()
```
